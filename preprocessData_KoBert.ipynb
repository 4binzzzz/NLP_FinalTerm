```python
import os
import json

def preprocess_data(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    filtered_utterances = []
    
    i = 0
    while i < len(data["utterances"]):
        current_utt = data["utterances"][i]
        
        # 현재 발화가 speaker이고 다음 발화가 listener인 경우
        if current_utt["role"] == "speaker" and (i + 1) < len(data["utterances"]):
            next_utt = data["utterances"][i + 1]
            if next_utt["role"] == "listener" and next_utt["listener_empathy"] is not None:
                # 공감 라벨이 있는 경우, 두 발화를 모두 추가
                filtered_utterances.append(current_utt)
                filtered_utterances.append(next_utt)
                i += 2  # 다음 두 발화로 이동
            else:
                i += 2  # 공감 라벨이 없는 경우, 다음 두 발화로 이동
        else:
            i += 1  # 현재 발화가 speaker가 아니면 다음 발화로 이동

    data["utterances"] = filtered_utterances

    return data

def process_directory(input_dir, output_dir):
    # 입력 디렉토리의 하위 디렉토리 및 파일을 반복
    for root, dirs, files in os.walk(input_dir):
        for file in files:
            if file.endswith(".json"):
                # 원본 파일 경로
                input_file_path = os.path.join(root, file)
                # 출력 파일 경로
                relative_path = os.path.relpath(root, input_dir)
                output_file_dir = os.path.join(output_dir, relative_path)
                os.makedirs(output_file_dir, exist_ok=True)
                output_file_path = os.path.join(output_file_dir, file)
                
                # 파일 전처리
                processed_data = preprocess_data(input_file_path)
                
                # 전처리된 데이터 저장
                with open(output_file_path, 'w', encoding='utf-8') as f:
                    json.dump(processed_data, f, ensure_ascii=False, indent=4)
                
                #print(f"Processed and saved: {output_file_path}")

# 원본 데이터 디렉토리
input_directory = 'D:/자연어처리/공감형 대화/Validation/02.라벨링데이터'
# 전처리된 데이터를 저장할 디렉토리
output_directory = 'D:/자연어처리/공감형 대화/Validation/VL_감정분석_전체데이터'

process_directory(input_directory, output_directory)
```
